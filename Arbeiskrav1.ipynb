```python
import pandas as pd

df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")
```

```python
#Task 1a

#Copy the values from dataframe
new_df = df[['Year','log q', 'log p', 'log h']].copy()


import numpy as np

#Making the original values for log p, log q and log h
new_df['P'] = np.exp(new_df['log p'])
new_df['Q'] = np.exp(new_df['log q'])
new_df['H'] = np.exp(new_df['log h'])

new_df
```

```python
#Task 1b
import numpy as np
from matplotlib import pyplot as plt

            
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(20,5))


#Plot the graphs
ax1.plot(new_df['Year'], new_df['H'], label='Harvest')
ax1.plot(new_df['Year'], new_df['Q'], label='Crop')
ax2.plot(new_df['Year'], new_df['P'], label='Price')

#Set the label for y axes
ax1.set(ylim=(0,150))

ax2.set(ylim=(0,450))

#Set the name for graphs
fig.suptitle('Watermelon price and quantities')
ax1.set_ylabel('Quantity')
ax1.set_xlabel('Year')

ax2.set_ylabel('Quantity')
ax2.set_xlabel('Year')

ax1.legend(loc='upper right')
ax2.legend(loc='upper right')
```

```python
#Task 2a, 2b and 2c

#Making a list for Q and logQ
Q = list(range(20,101,5))
Log_Q = list(np.log(Q))

#Making a supply and demand equation in log
Supply_logP = [ -2+1.7*x for x in Log_Q ] 
Demand_logP = [ 8.5 - 0.82*x for x in Log_Q]

#Turning the supply and the demand in to their original form
Supply_p = np.exp(Supply_logP)
Demand_p = np.exp(Demand_logP)


from tabulate import tabulate

#Plotting all the list in the right spaces
print(tabulate({'Q': Q, 
                'log Q': Log_Q, 
                'Supply(log P)': Supply_logP,
               'Demand(log P)': Demand_logP,
               'Supply(P)':Supply_p,
               'Demand(P)':Demand_p}, 
                headers='keys', 
                tablefmt='fancy_grid', 
                missingval='N/A'))
```

```python
#Task 2d

#Making x-axis
x = np.linspace(0,100,17)


fig,ax=plt.subplots()


#Giving the labels names
ax.set_ylabel('Price, P(US$ per 1,000 watermelons)')
ax.set_xlabel('Quantity, Q(millions)')


#Plotting the supply and demand
ax.plot(x,Supply_p,label='Supply')
ax.plot(x,Demand_p, color="red", label='Demand')


#Setting the axes
ax.set(xlim=(0,100))
ax.set(ylim=(0,450))



ax.set_title('Supply and demand diagram.')
ax.set_ylabel('Price, P(US$ per 1,000 watermelons)')
ax.set_xlabel('Quantity, Q(millions)')

#legend:
ax.legend(loc='upper right')


plt.show()
```

```python
#Task 3a:

#Making the new supply
New_Supply_logP = [ -2+1.7*x+0.4 for x in Log_Q ] 
New_Supply_p = np.exp(New_Supply_logP)

print(tabulate({'Q': Q, 
                'log Q': Log_Q, 
                'Supply(log P)': Supply_logP,
               'Demand(log P)': Demand_logP,
               'Supply(P)':Supply_p,
               'Demand(P)':Demand_p,
               'New Supply(log P)' : New_Supply_logP,
               'New Supply(P)' : New_Supply_p},
                headers='keys', 
                tablefmt='fancy_grid', 
                missingval='N/A'))
```

```python
#Task 3b

#Making x-axis
x = np.linspace(0,100,17)


fig,ax=plt.subplots()


#Plotting the supply, demand and the new supply
ax.plot(x,Supply_p,label='Supply')
ax.plot(x,Demand_p, color="red", label='Demand')
ax.plot(x,New_Supply_p, color="green", label='New Supply')



ax.set(xlim=(0,100))
ax.set(ylim=(0,450))


#Giving the labels names
ax.set_title('Supply and demand diagram.')
ax.set_ylabel('Price, P(US$ per 1,000 watermelons)')
ax.set_xlabel('Quantity, Q(millions)')

#legend:
ax.legend(loc='upper right')


plt.show()
```
